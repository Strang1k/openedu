PCA - метод главных компонент (principal component analysis), снижение размерности
Поиск вложеных многообразий; линия, плоскость, отртонормированный набор векторов

Данный метод аппроксимирует n-размерное облако наблюдений до эллипсоида (тоже n-мерного), полуоси которого и будут являться будущими главными компонентами. И при проекции на такие оси (снижении размерности) сохраняется наибольшее количество информации.

В основе линейная алгебра, синий вектор, красные, разность синего вектора (хi - а0) и суммы проекций до осей гиперплоскости

from sklearn.decomposition import PCA
pca = PCA(n_components = 1)
XPCAreduced = pca.fit_transform(transpose(X))

t-SNE
Нелинейные преобразования для визуализации данных, позволяют из многомерного привести к 2-3мерному

Распределение расстояний между точками похоже на изначальное

При изменении гиперпараметра (учет соседей) получается разная картина которую можно поразному интерпретировать

Для новой точки нужно все заново переписывать

Перплексия подбирается бинарным поиском от 5 до 50 (оптимально)

Желательно знать хоть что-то о данных от чего отталкиваться

Качество определяется тем, насколько она помогла решить проблему
